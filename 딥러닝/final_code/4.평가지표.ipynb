```python
import pandas as pd

df = pd.DataFrame(columns = ['FID', "Mean PSNR", "Mean SSIM", "Mean MS_SSIM"],
                        index = [f"{i}epoch" for i in range(1, 41)])

for i in range(1, 41):
  f = open(f"/content/drive/MyDrive/딥러닝/최종코드/yejin_ChromaGAN/Output_final/{i}epoch_metrics.txt", "r")
  dt = f.read().splitlines()
  df.loc[f"{i}epoch"] = [float(i.split(" ")[-1]) for i in dt]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 1, figsize = (10, 20))
df[["FID", "Mean PSNR"]].plot(ax = ax[0])
plt.legend(fontsize = 15, loc = "upper left")
df[["Mean SSIM", "Mean MS_SSIM"]].plot(ax =ax[1])
plt.legend(fontsize = 15, loc = "upper left")
```
